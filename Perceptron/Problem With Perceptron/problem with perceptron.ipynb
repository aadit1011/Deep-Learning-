```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from mlxtend.plotting import plot_decision_regions
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input1']=[1,1,0,0]
or_data['input2']=[1,0,1,0]
or_data['output']=[1,1,1,0]
```

```python
or_data 
```

```python
and_data['input1']=[1,1,0,0]
and_data['input2']=[1,0,1,0]
and_data['output']=[1,0,0,0]
```

```python
and_data
```

```python
xor_data['input1']=[1,1,0,0]
xor_data['input2']=[1,0,1,0]
xor_data['output']=[0,1,1,0]
```

```python
xor_data
```

```python
sns.scatterplot(data=and_data,x='input1',y='input2',hue='output')
```

```python
sns.scatterplot(data=or_data,x='input1',y='input2',hue='output')
```

```python
sns.scatterplot(data=xor_data,x='input1',y='input2',hue='output')
```

```python
from sklearn.linear_model import Perceptron
clf1=Perceptron()
clf2=Perceptron()
clf3=Perceptron()
```

```python
clf1.fit(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values)
```

```python

clf2.fit(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values)
```

```python

clf3.fit(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values)
```

```python
plot_decision_regions(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values, clf=clf1, legend=2)
plt.savefig('decision_boundary_and_using_perceptron.png')
```

```python
clf1.coef_
```

```python
w1, w2 = clf1.coef_[0]
```

```python
w1 
```

```python
w2 
```

```python
c1=clf1.intercept_
c1 
```

```python
c1_new=c1[0]
c1_new
```

```python
and_data['input1']
```

```python
x_new1=np.unique(and_data['input1'])
x_new1
```

```python
y_new1=((-w1/w2)*x_new1)-(c1_new/w2)
y_new1
```

```python
y_new1
```

```python
plt.plot(x_new1,y_new1,color='red') 
sns.scatterplot(x=and_data['input1'],y=and_data['input2'],hue=and_data['output'])
plt.savefig('line_plot_and_data.png')
plt.show()
```

```python
plot_decision_regions(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values, clf=clf2, legend=2)
plt.savefig('decision_boundary_or_data_using_perceptron.png')
```

```python
w1_clf2,w2_clf2=clf2.coef_[0]
```

```python
c2=clf2.intercept_[0]
c2 
```

```python
w1_clf2
```

```python
w1_clf2
```

```python
x_new2=np.unique(or_data['input1'])
x_new2
```

```python
y_new2=(-1*(w1_clf2/w2_clf2)*x_new2)-(c2/w2_clf2)
y_new2
```

```python
plt.plot(x_new2,y_new2,color='red') 
sns.scatterplot(x='input1',y='input2',hue='output',data=or_data)
plt.savefig('line_plot_or_data.png')
plt.show()
```
