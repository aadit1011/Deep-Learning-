```python
import numpy as np

def initialize_parameters():

    params = {
        'W1': np.random.randn(2, 4) * 0.01,
        'b1': np.zeros((2, 1)),
        'W2': np.random.randn(2, 2) * 0.01,
        'b2': np.zeros((2, 1)),
        'W3': np.random.randn(1, 2) * 0.01,
        'b3': np.zeros((1, 1))
    }
    return params

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def forward_propagation(X, params):
    # Layer 1
    Z1 = np.dot(params['W1'], X) + params['b1']
    A1 = np.tanh(Z1)
    # Layer 2
    Z2 = np.dot(params['W2'], A1) + params['b2']
    A2 = np.tanh(Z2)
    # Layer 3 (Output)
    Z3 = np.dot(params['W3'], A2) + params['b3']
    A3 = sigmoid(Z3)

    cache = {"Z1": Z1, "A1": A1, "Z2": Z2, "A2": A2, "Z3": Z3, "A3": A3}
    return A3, cache

def back_propagation(X, Y, cache, params):
    m = X.shape[1]

    # Output layer
    dZ3 = cache['A3'] - Y
    dW3 = (1/m) * np.dot(dZ3, cache['A2'].T)
    db3 = (1/m) * np.sum(dZ3, axis=1, keepdims=True)

    # Hidden layer 2
    dA2 = np.dot(params['W3'].T, dZ3)
    dZ2 = dA2 * (1 - np.power(cache['A2'], 2))
    dW2 = (1/m) * np.dot(dZ2, cache['A1'].T)
    db2 = (1/m) * np.sum(dZ2, axis=1, keepdims=True)

    # Hidden layer 1
    dA1 = np.dot(params['W2'].T, dZ2)
    dZ1 = dA1 * (1 - np.power(cache['A1'], 2))
    dW1 = (1/m) * np.dot(dZ1, X.T)
    db1 = (1/m) * np.sum(dZ1, axis=1, keepdims=True)

    grads = {"dW1": dW1, "db1": db1, "dW2": dW2, "db2": db2, "dW3": dW3, "db3": db3}
    return grads

def update_parameters(params, grads, learning_rate=0.01):
    params['W1'] -= learning_rate * grads['dW1']
    params['b1'] -= learning_rate * grads['db1']
    params['W2'] -= learning_rate * grads['dW2']
    params['b2'] -= learning_rate * grads['db2']
    params['W3'] -= learning_rate * grads['dW3']
    params['b3'] -= learning_rate * grads['db3']
    return params
```

```python

X_data = x.T
Y_data = (y == 0).reshape(1, -1)

parameters = initialize_parameters()
epochs = 1000
learning_rate = 0.1
total_cost=0
epoch_cost=0

for i in range(epochs):
    # Forward
    A3, cache = forward_propagation(X_data, parameters)

    # Cost (Binary Cross Entropy)
    cost = -np.mean(Y_data * np.log(A3) + (1 - Y_data) * np.log(1 - A3))

    # Backward
    grads = back_propagation(X_data, Y_data, cache, parameters)

    # Update
    parameters = update_parameters(parameters, grads, learning_rate)
    cost = -np.mean(Y_data * np.log(A3 + 1e-8) + (1 - Y_data) * np.log(1 - A3 + 1e-8))
    epoch_cost += cost
    total_cost+=cost


    if i % 100 == 0:
        print(f"Epoch {i}, Avg Epoch Cost: {epoch_cost / X_data.shape[0]}")
print(f'Total Cost is {total_cost}')
print(f"Epoch {i}, Avg Total Cost: {total_cost / (X_data.shape[0]*epochs)}")
```

Mini Batch Gradient Descent


```python
X_data = x
Y_data = (y == 0).astype(int)

parameters = initialize_parameters()
epochs = 1000
learning_rate = 0.01
total_cost=0
batch_size=50
for i in range(epochs):
    epoch_cost = 0
    num1=0
    num2=0
    while num1<=X_data.shape[0] and (X_data.shape[0]-num2)!=batch_size:
      num2=num1
      num1+=batch_size

      for j in range(num2,num1):
        xj = X_data[j].reshape(-1, 1)
        yj = Y_data[j].reshape(1, 1)

        # Forward
        A3, cache = forward_propagation(xj, parameters)

        # Backward
        grads = back_propagation(xj, yj, cache, parameters)

        # Update
        parameters = update_parameters(parameters, grads, learning_rate)

        cost = -np.mean(yj * np.log(A3 + 1e-8) + (1 - yj) * np.log(1 - A3 + 1e-8))
        epoch_cost += cost
        total_cost+=cost

    if i % 100 == 0:
        print(f"Epoch {i}, Avg Epoch Cost: {epoch_cost / X_data.shape[0]}")
print(f'Total Cost is {total_cost}')
print(f"Epoch {i}, Avg Total Cost: {total_cost / (X_data.shape[0]*epochs)}")
```
