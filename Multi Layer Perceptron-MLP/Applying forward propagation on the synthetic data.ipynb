# Defining  Sigmoid Function 

```python
def sigmoid(z): 
    return 1/(1+np.exp(-z))
```

# For the first hidden Layer 

```python
np.random.seed(42)
w1=np.random.randn(input_columns,hidden1)*0.4
b1=np.random.randn(1,hidden1)
w1 
```

```python
w1.shape
```

```python
b1 
```

```python
b1.shape
```

```python
X.shape
```

```python
dot_hidden1=np.dot(X,w1)
```

```python
dot_hidden1.shape
```

```python
output_hidden1=dot_hidden1+b1
```

```python
output_hidden1_acf=sigmoid(output_hidden1)
```

```python
output_hidden1_acf.shape
```

# Now Output From Hidden 1 becomes input to the hidden layer 2 

```python
input_hidden2=output_hidden1_acf
```

```python
input_hidden2.shape
```

```python
output_hidden1_acf
```

# weights and biases for the hidden layer 2  

```python
w2=np.random.randn(hidden1,hidden2)*0.4
b2=np.random.randn(1,hidden2)
w2 
```

```python
dot_hidden2=np.dot(output_hidden1_acf,w2)
output_hidden2=dot_hidden2+b2
```

```python
output_hidden2_acf=sigmoid(output_hidden2)
output_hidden2_acf
```

```python
input_output_layer=output_hidden2_acf
```

```python
input_output_layer.shape
```

```python
w3=np.random.randn(hidden2,output)*0.4
b3=np.random.randn(1,output)
w3 
```

```python
dot_hidden3=np.dot(input_output_layer,w3)
output_output_layer=dot_hidden3+b3
output_output_layer_acf=sigmoid(output_output_layer)
output_output_layer_acf
```
